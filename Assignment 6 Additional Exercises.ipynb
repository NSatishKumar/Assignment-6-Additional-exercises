## 5. Calculate descriptive statistics for the 'ratings' column of the ratings DataFrame. Verify using describe().

```python
print("Statistics for rating column in Ratings Dataframe")
print(df_ratings["rating"].describe())                          # statistics of rating column in  rating dataset
print("\n")
print("count: ",df_ratings["rating"].count())                          # count of rating column in  rating dataset
print("mean: ",df_ratings["rating"].mean())                            # mean of rating column in  rating dataset
print("minimum value: ",df_ratings["rating"].min())                    # minimum of rating column in  rating dataset
print("maximum value: ",df_ratings["rating"].max())                    # maximum of rating column in  rating dataset
print("median value: ",df_ratings["rating"].median())                  # median of rating column in  rating data
print("mode value: ",df_ratings["rating"].mode())                      # mode of rating column in  rating dataset
print("25% value: ",df_ratings["rating"].quantile(0.25))               # 25% of rating column in  rating dataset
print("50% value: ",df_ratings["rating"].quantile(0.50))               # 50% of rating column in  rating data
print("75% value: ",df_ratings["rating"].quantile(0.75))               # 75% of rating column in  rating data
print("standard deviation value: ",df_ratings["rating"].std())         # standard deviation of rating column in  rating data
```

## 7. Find how many null values, missing values are present. Deal with them. Print out how many rows have been modified. 


```python
print("nulls values information in ratings")
ratingNulls=df_ratings.isna().sum()                             # Calculating the null values in each column of rating dataframe
print(ratingNulls)

print("\nnulls values information in tags")
tagsNulls=df_tags.isna().sum()                                 # Calculating the null values in each column of tags dataframe
print(tagsNulls)
tagsshapeb4=df_tags.shape
print("\nnulls values information in movies")
moviesNulls=df_movies.isna().sum()                             # Calculating the null values in each column of Movies dataframe
print(moviesNulls)

print("number of null values in ratings are ",ratingNulls.sum())  # Dsiplaying the total count of null values in rating dataframe
print("number of null values in tags are ",tagsNulls.sum())       # Dsiplaying the total count of null values in tags dataframe
print("number of null values in movies are ",moviesNulls.sum())   # Dsiplaying the total count of null values in movies dataframe

df_tags = df_tags.dropna()                                        # Dropping the null values present in the tags
tagsshapebaftr=df_tags.shape
print("number of rows modified in tags are ",int(tagsshapeb4[0])-int(tagsshapebaftr[0]))     # Numebr of rows modified in the tags dataframe
  
```

## 8. Filter out movies from the movies DataFrame that are of type 'Animation'.

## 9. Find the average rating of movies. 

```python
df_avg = df_ratings.groupby('movieId').mean()            # grouping the ratings data frame basedon movieID and calculating the mean/average
df_ratingCol=df_avg["rating"]                            # Storing the rating column into df_ratingCol
df_avgMovies=pd.DataFrame(df_ratingCol)                  # Converting into dataframe
df_avgMovies.sort_values("rating",ascending=False)       # Sorting the movies based in descending order
```

## 11. Print out the 5 movies that belong to the Comedy genre and have rating greater than 4.

```python
df_avg = df_ratings.groupby('movieId').mean()                       # grouping the ratings data frame basedon movieID and calculating the mean/average
df_AvgGr4=df_avg[df_avg["rating"]>4]                                # Filtering the data based upon condition rating > 4
df_temp=df_movies.merge(df_AvgGr4, on='movieId', how='inner')       # inner join on movies and above dataframe
df_ComedyGr4=df_temp[df_temp["genres"].str.contains('Comedy',case = False)]  # filtering the data frame based on genre= Comedy
df_ComedyGr4.head()
```

```python
df_movies['year'] = df_movies['title'].str.extract('.*\((.*)\).*', expand=True)   # Extraing the year from movie title
df_movies
```

## 14. Select rows based on timestamps later than 2015-02-01.

```python
df_tags["Date"]=pd.to_datetime(df_tags['timestamp'], unit='s')   # Converting the timestamp into date time  and storing in a new column
df_timestamps= df_tags[df_tags["Date"]>"2015-02-01"]             # Filtering the data basedon condition >"2015-02-01"
df_timestamps
```
